```python
from tkinter import *
import random
```

```python
# menambahkan variable global

# variable global
WIDTH = 900
HEIGHT = 300

# pengaturan raket

# lebar raket
PAD_W = 10
# tinggi raket
PAD_H = 100

# pengaturan bola
# seberapa besar kecepatan bola akan meningkat dengan setiap puukuklan
BALL_SPEED_UP = 1.05
# kecepatan bola maximum
BALL_MAX_SPEED = 40
# radius bola
BALL_RADIUS = 30

INITIAL_SPEED = 20
BALL_X_SPEED = INITIAL_SPEED
BALL_Y_SPEED = INITIAL_SPEED

# skor pemain
PLAYER_1_SCORE = 0
PLAYER_2_SCORE = 0

# menambahkan variable global untuk jarak
# ke tepi kanan lapangan permainan
right_line_distance = WIDTH - PAD_W 
```

```python
def update_score(player):
  global PLAYER_1_SCORE, PLAYER_2_SCORE
  if player == "right":
    PLAYER_1_SCORE += 1
    c.itemconfig(p_1_text, text=PLAYER_1_SCORE)
  else:
    PLAYER_2_SCORE += 1
    c.itemconfig(p_2_text, text=PLAYER_2_SCORE)
```

```python
def spawn_ball():
  global BALL_X_SPEED
  # letakan bola di bagian tengah
  c.coords(BALL,
           WIDTH/2-BALL_RADIUS/2,
           HEIGHT/2-BALL_RADIUS/2,
           WIDTH/2+BALL_RADIUS/2,
           HEIGHT/2+BALL_RADIUS/2)
  # atur arah bola ke arah pemain yang kalah,
  # tetapi mengurangi kecepatan ke kecepatan semula
  BALL_X_SPEED = -(BALL_X_SPEED * -INITIAL_SPEED) / abs(BALL_X_SPEED)
```

```python
# fungsi bola pantul
def bounce(action):
  global BALL_X_SPEED, BALL_Y_SPEED
  # memukul dengan raket
  if action == "strike":
    BALL_Y_SPEED = random.randrange(-10, 10)
    if abs(BALL_X_SPEED) < BALL_MAX_SPEED:
      BALL_X_SPEED *= -BALL_SPEED_UP
    else:
      BALL_X_SPEED = -BALL_X_SPEED
  else:
    BALL_Y_SPEED = -BALL_Y_SPEED
```

```python
# memasang jendela
root = Tk()
root.title("Pong")

# area animasi
c = Canvas(root, width=WIDTH, height=HEIGHT, background="#003300")
c.pack()
```

```python
# elemen lapangan bermain

# garis kiri
c.create_line(PAD_W, 0, PAD_W, HEIGHT, fill="#fff")
# garis kanan
c.create_line(WIDTH-PAD_W, 0, WIDTH-PAD_W, HEIGHT, fill="#fff")
# garis tengah
c.create_line(WIDTH/2, 0, WIDTH/2, HEIGHT, fill="#fff")
```

```python
# pemasangan fasilitas permainan

# membuat bola
BALL = c.create_oval(WIDTH/2-BALL_RADIUS/2,
                     HEIGHT/2-BALL_RADIUS/2,
                     WIDTH/2+BALL_RADIUS/2,
                     HEIGHT/2+BALL_RADIUS/2, fill="#fff")

# raket kiri
LEFT_PAD = c.create_line(PAD_W/2, 0, PAD_W/2, PAD_H, width=PAD_W, fill="yellow")

# raket yang tepat
RIGHT_PAD = c.create_line(WIDTH-PAD_W/2, 0, WIDTH-PAD_W/2,
                          PAD_H, width=PAD_W, fill="yellow")

p_1_text = c.create_text(WIDTH-WIDTH/6, PAD_H/4,
                         text=PLAYER_1_SCORE,
                         font="Arial 20",
                         fill="white")

p_2_text = c.create_text(WIDTH/6, PAD_H/4,
                         text=PLAYER_2_SCORE,
                         font="Arial 20",
                         fill="white")
```

```python
# menambahkan variable global untuk kecepatan bola
# di seluruh
BALL_X_CHANGE = 20
# secara vertikal
BALL_Y_CHANGE = 0
```

```python
def move_ball():
  # menentukan kordinat sisi-sisi bola dan pusatnya
  ball_left, ball_top, ball_right, ball_bot = c.coords(BALL)
  ball_center = (ball_top + ball_bot) / 2

  # pantulan vertikal
  # jika kita berada jauh dari garis vertikal - cukup gerakkan bola
  if ball_right + BALL_X_SPEED < right_line_distance and \
          ball_left + BALL_X_SPEED > PAD_W:
          c.move(BALL, BALL_X_SPEED, BALL_Y_SPEED)
  # jika bola menyentuh sisi kanan atau kiri batas lapangan
  elif ball_right == right_line_distance or ball_left == PAD_W:
    # memeriksa sisi kanan atau kiri yang kita sentuh
    if ball_right > WIDTH / 2:
      # jika benar, bandingkan posisi bagian tengah bola
      # dengan posisi raket yang tepat.
      # dan jika bola berada di dalam raket, lakukan pantulan
      if c.coords(RIGHT_PAD)[1] < ball_center < c.coords(RIGHT_PAD)[3]:
        bounce("strike")
      else:
        # jika tidak, seorang pemain akan kehilangn keesempatan - di sini kita tinggalkan operan untuk
        update_score("left")
        spawn_ball()
    else:
    # sama untuk pemain kidal
      if c.coords(LEFT_PAD)[1] < ball_center < c.coords(LEFT_PAD)[3]:
        bounce("strike")
      else:
        update_score("right")
        spawn_ball()
# periksa situasi di mana bola dapat terbang keluar dari batas lapangan permainan.
# dalam hal ini, cukup pindahkan ke tepi bidang.
  else:
    if ball_right > WIDTH / 2:
      c.move(BALL, right_line_distance-ball_right, BALL_Y_SPEED)
    else:
      c.move(BALL, -ball_left+PAD_W, BALL_Y_SPEED)
# pantulan horizontal
  if ball_top + BALL_Y_SPEED < 0 or ball_bot + BALL_Y_SPEED > HEIGHT:
    bounce("ricochet")
```

```python
# mengatur variable global untuk kecepatan raket
# kecepatan gerak raket yang akan dilalui
PAD_SPEED = 20
# kecepatan plaatform kiri
LEFT_PAD_SPEED = 0
# kecepatan raket yang tepat
RIGHT_PAD_SPEED = 0
```

```python
# fungsi pergrakan kedua raket
def move_pads():
  # untuk kenyamanan, mari kita ciptakan kosakataa di mana raket sesuai dengan kecepatanya
  PADS = {LEFT_PAD: LEFT_PAD_SPEED,
          RIGHT_PAD: RIGHT_PAD_SPEED}
  for pad in PADS:
    # mengerrakan raket dengan kecepatan tertentu
    c.move(pad, 0, PADS[pad])
    # jika raket bergerak keluar dari bidang permainan, letakan kembali pada tempatnya
    if c.coords(pad)[1] < 0:
      c.move(pad, 0, -c.coords(pad)[1])
    elif c.coords(pad)[3] > HEIGHT:
      c.move(pad, 0, HEIGHT - c.coords(pad)[3])
```

```python
def main():
  move_ball()
  move_pads()
  # memanggil dirinya sendiri setiap 30 milidetik
  root.after(30, main)
```

```python
# mengatur fokus pada canvas untuk merespons penekanan tombol
c.focus_set()
```

```python
# mari menulis fungsi untuk menangani penekanan tombol
def movement_handler(event):
  global LEFT_PAD_SPEED, RIGHT_PAD_SPEED
  if event.keysm == "w":
    LEFT_PAD_SPEED = -PAD_SPEED
  elif event.keysym == "s":
    LEFT_PAD_SPEED = PAD_SPEED
  elif event.keysym == "Up":
    RIGHT_PAD_SPEED = -PAD_SPEED
  elif event.keysym == "Down":
    RIGHT_PAD_SPEED = PAD_SPEED
```

```python
# bind this function to canvass
c.bind("<KeyPress>", movement_handler)
```

```python
# create a button release response function
def stop_pad(event):
  global LEFT_PAD_SPEED, RIGHT_PAD_SPEED
  if event.keysym in "ws":
    LEFT_PAD_SPEED = 0
  elif event.keysym in ("Up", "Down"):
    RIGHT_PAD_SPEED = 0
```

```python
#  bind this function to canvas
c.bind("<KeyRelease>", stop_pad)
```

```python
# pengaturan dalam gerakan
main()
```

```python
# menjalankan jendela
root.mainloop()
```
